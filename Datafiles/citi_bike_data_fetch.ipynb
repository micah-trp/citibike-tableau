```python
###########################
## CREATE trip_distance

def calculate_distance(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = 6371 * c  # Radius of the Earth in kilometers
    return distance
```

```python
################################################

### Create table of Stations - create as start and end with coordinates- concatinate then delete any duplicates.
# 2019
jun_19_start = jul_19_pd[['start station name',
'start station latitude',
'start station longitude']]

jun_19_end = jul_19_pd[['end station name',
'end station latitude',
'end station longitude' ]]

#--
jul_19_start = jul_19_pd[['start station name',
'start station latitude',
'start station longitude']]

jul_19_end = jul_19_pd[['end station name',
'end station latitude',
'end station longitude' ]]

#--
aug_19_start = aug_19_pd[['start station name',
'start station latitude',
'start station longitude']]

aug_19_end = aug_19_pd[['end station name',
'end station latitude',
'end station longitude']]

# 2020
jun_20_start = jul_20_pd[['start station name',
'start station latitude',
'start station longitude']]

jun_20_end = jul_20_pd[['end station name',
'end station latitude',
'end station longitude' ]]

#--
jul_20_start = jul_20_pd[['start station name',
'start station latitude',
'start station longitude']]

jul_20_end = jul_20_pd[['end station name',
'end station latitude',
'end station longitude' ]]

#--
aug_20_start = aug_20_pd[['start station name',
'start station latitude',
'start station longitude']]

aug_20_end = aug_20_pd[['end station name',
'end station latitude',
'end station longitude']]

```

```python

stations_start_raw_all = pd.concat([jun_19_start, jul_19_start,aug_19_start,jun_20_start,jul_20_start,aug_20_start], axis=0)
stations_end_raw_all = pd.concat([jun_19_end, jul_19_end,aug_19_end,jun_20_end,jul_20_end,aug_20_end], axis=0)

# Group the DataFrame by the specified columns
start_grouped = stations_start_raw_all.groupby(['start station name',
                         'start station latitude', 'start station longitude']).size().reset_index()


end_grouped = stations_end_raw_all.groupby(['end station name',
                         'end station latitude', 'end station longitude']).size().reset_index()

# Rename the count column
start_grouped = start_grouped.drop(columns=[0])
start_grouped = start_grouped.drop_duplicates()


end_grouped = end_grouped.drop(columns=[0])
end_grouped = end_grouped.drop_duplicates()

```

```python
#############
## Rename columns the fit into one concat df and remove duplicates

start_grouped = start_grouped.rename(columns={'start station name': 'station name','start station latitude': 'station latitude','start station longitude': 'station longitude'})
start_grouped_df = pd.DataFrame(start_grouped).reset_index()



end_grouped = end_grouped.rename(columns={'end station name': 'station name','end station latitude': 'station latitude','end station longitude': 'station longitude'})
end_grouped_df = pd.DataFrame(end_grouped).reset_index()
```

```python
stations = pd.concat([start_grouped_df,end_grouped_df], axis=0)
stations_df = pd.DataFrame(stations.drop_duplicates())

stations_df = stations_df.reset_index(drop=True)
stations_df.info()
```
